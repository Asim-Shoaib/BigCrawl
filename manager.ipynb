```python
import os 
import json
```

```python
url_map_path = os.path.join(os.getcwd(), 'urls_data', 'url_map.json')
url_map = {}
with open(url_map_path, 'r') as f:
    url_map = json.load(f)
```

```python
no_metmuseum_art_catalog = 0
to_delete = []
for key, value in url_map.items():
    if 'metmuseum.org/art/collection/search' in value:
        no_metmuseum_art_catalog += 1
        to_delete.append(key)
```

```python
for key in to_delete:
    # print(f"Deleting key: {key} with URL: {url_map[key]}")
    del url_map[key]
    delete_file_path = os.path.join(os.getcwd(), 'urls_data', 'raw', f'{key}.html')
    if os.path.exists(delete_file_path):
        # print(f"Deleting {delete_file_path}")
        os.remove(delete_file_path)
```

```python
with open(url_map_path, 'w') as f:
    json.dump(url_map, f, indent=2)
```

```python
from bs4 import BeautifulSoup
def is_page_english_by_metadata(html):
    soup = BeautifulSoup(html, "html.parser")

    # Check <html lang="en">
    html_tag = soup.find("html")
    if html_tag:
        lang = html_tag.get("lang") or html_tag.get("xml:lang")
        print(lang)
        if lang and lang.lower().startswith("en"):
            return True
        else:
            return False

    # Check <meta http-equiv="content-language">
    meta = soup.find("meta", attrs={"http-equiv": "content-language"})
    if meta:
        lang = meta.get("content", "").lower()
        print(lang)
        if "en" in lang:
            return True
        else:
            return False

    # No definitive metadata â†’ return None (unknown)
    return None
```

```python
import random
html = ""
ts = 0
fs = 0
for i in range(1, 10):
    with open(os.path.join(os.getcwd(), 'urls_data', 'raw', f"{list(url_map.keys())[random.randint(1,10000)]}.html"), 'r', encoding='utf-8') as f:
        html = f.read()
    res = is_page_english_by_metadata(html)
    if res == True:
        ts += 1
    elif res == False:
        fs += 1
```
