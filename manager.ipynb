```python
import os 
import json
```

```python
url_map_path = os.path.join(os.getcwd(), 'urls_data', 'url_map.json')
url_map = {}
with open(url_map_path, 'r') as f:
    url_map = json.load(f)
```

```python
no_metmuseum_art_catalog = 0
to_delete = []
for key, value in url_map.items():
    if 'metmuseum.org/art/collection/search' in value:
        no_metmuseum_art_catalog += 1
        to_delete.append(key)
```

```python
for key in to_delete:
    # print(f"Deleting key: {key} with URL: {url_map[key]}")
    del url_map[key]
    delete_file_path = os.path.join(os.getcwd(), 'urls_data', 'raw', f'{key}.html')
    if os.path.exists(delete_file_path):
        # print(f"Deleting {delete_file_path}")
        os.remove(delete_file_path)
```

```python
with open(url_map_path, 'w') as f:
    json.dump(url_map, f, indent=2)
```
